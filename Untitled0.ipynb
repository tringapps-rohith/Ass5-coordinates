```python
import numpy as np
import pandas as pd
df=pd.read_csv("/housing.csv")
df.head()
df=df.dropna()
df.count()
```

```python
from statsmodels.tsa.stattools import adfuller
test_result=adfuller(df['median_house_value'])
test_result
```

```python
def adfuller_test(median_house_value):
    result=adfuller(median_house_value)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )

adfuller_test(df['median_house_value'])
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df['median_house_value'])
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(df['median_house_value'],order=(1,1,1))
model_fit=model.fit()
model_fit.summary()
```

```python
import statsmodels.api as sm
model=sm.tsa.statespace.SARIMAX(df['median_house_value'],order=(1, 1, 1),seasonal_order=(1,1,1,12))
results=model.fit()
df['forecast']=results.predict()
df['forecast']=df['forecast'][16346:]
df['median_house_value']=df['median_house_value'][16346:]
df[['median_house_value','forecast']].plot(figsize=(12,8))
```

```python
df=df[['median_house_value','forecast']]
df[16346:]
```

```python
ytest=df['median_house_value']
ypred=df['forecast']
mape=np.mean(np.abs((ytest-ypred)/ytest)*100)
100-mape
```
